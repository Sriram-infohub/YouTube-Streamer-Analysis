```python
#Trend Analyisis
"""Identify trends among the top YouTube streamers. Which categories are the most popular?
   Is there a correlation between the number of subscribers and the number of likes or comments?"""
plt.figure(figsize=(12,8))
x=df.groupby('Categories')['Suscribers'].sum().plot(kind='bar')
plt.title("Category vs Subscribers")
```

```python
df1=df[['Suscribers','Likes','Comments']]
sns.heatmap(df1.corr(),annot=True)
```

```python
#Audience Study:
""" - Analyze the distribution of streamers' audiences by country. Are there regional preferences for
     specific content categories?"""

country_category = df.groupby(['Country', 'Categories'])['Rank'].count().reset_index(name='Count')

fig = px.treemap(country_category, path=['Country', 'Categories'], values='Count',
                  title='Distribution of Streamers by Country and Category')
fig.show()
```

```python
#Performance Metrics:
"""- Calculate and visualize the average number of subscribers, visits, likes, and comments.
   - Are there patterns or anomalies in these metrics?"""


df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean().plot(kind='bar')
plt.title('Average Metrics')
plt.ylabel('Average')
```

```python
#Content Categories:
"""- Explore the distribution of content categories. Which categories have the highest number of streamers?
   - Are there specific categories with exceptional performance metrics"""

category_counts = df['Categories'].value_counts()
print(category_counts)

category_metrics = df.groupby('Categories')[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
print(category_metrics)

overall_average = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
exceptional_categories = category_metrics[category_metrics > overall_average]
print(exceptional_categories)

```

```python
#Brands and Collaborations:
"""- Analyze whether streamers with high performance metrics receive more
     brand collaborations and marketing campaigns."""

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Suscribers', y='Links', data=df)
plt.title('Subscribers vs Brand Collaborations')
plt.xlabel('Subscribers')
plt.ylabel('Brand Collaborations')
plt.show()


```

```python
 #Benchmarking:
"""- Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments.
   - Who are the top-performing content creators?"""

avg_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
above_average = df[['Suscribers', 'Visits', 'Likes', 'Comments']] > avg_metrics
top_performers = df[above_average.all(axis=1)]
print(top_performers)
```
